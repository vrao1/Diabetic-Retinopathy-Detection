```python
import pandas as pd
from pandas import DataFrame
import numpy as np
import matplotlib.pyplot as plt
import nltk
from collections import Counter, OrderedDict
from pandas import compat
```

```python
table_names = ['all_encounter_data', 'demographics', 'encounters', 'family_hist_for_Enc',
 'family_hist_list', 'ICD_for_Enc', 'macula_findings_for_Enc',
 'SL_Lens_for_Enc', 'SNOMED_problem_list', 'systemic_disease_for_Enc', 'systemic_disease_list']

data_analysis_index = [5,4,0,5,3,5,5,5,3,5,3]
da_index = 0
top_N = 30
```

```python
PATH = '/home/vinodrao/Spring2017/MATH497/project/DataAnalysis/'

dfs = [pd.read_pickle(PATH + name + '.pickle') for name in table_names]

def dict_to_df(d):
    #df=pd.DataFrame()
    #df['words'] = d.keys()
    #df['Frequency'] = d.values()
    print(d.keys())
    print(d.values())
    df = pd.DataFrame(d.values())
    #df.set_index(0, inplace=True)
    return df

def to_dict_dropna(data):
  return dict_to_df(dict((k, v.dropnan().to_dict()) for k, v in compat.iteritems(data)))
```

```python
for index,item in enumerate(table_names):
    print "For Table " + item
    print "================================================="
    num = 1
    df = dfs[index].replace(r'^\s+','', regex=True).replace(r'\s+$','', regex=True).apply(lambda x: x.astype(str).str.lower())
    df_tmp = df.replace(r'', np.nan, regex=True)
    #df_tmp.apply(lambda y: pd.Series(y).dropna(inplace=True))
    for i, row in  df_tmp.T.iterrows():
        print "Column [" + str(num) + "] = '" + i + "'"
        print "\t*******************************************"
        print "\t Length = " + str(len(row))   
        print "\t Total NaN values = " + str(row.isnull().sum())
        print "\t Total Unique Values = " + str(len(row.unique()))
        if(num > data_analysis_index[da_index]):
            word_dist = nltk.FreqDist(row)
            #print(type(word_dist))
            rslt = pd.DataFrame(word_dist.most_common(top_N), columns=['Word', 'Frequency'])
            #rslt = pd.DataFrame(word_dist, columns=['Word', 'Frequency'])
            print('-' * 60)
            print("Most Frequent Keywords in column '" + i + "'")
            print('-' * 60)
            #print(type(rslt))
            print(rslt)
            if(da_index != 2):
                print('-' * 60)
                print("Histogram to show distribution of data under column '" + i + "'")
                print('-' * 60)
                #print(type(row))
                col = row.tolist()
                cleanedList = [x for x in col if str(x) != 'nan']
                letter_counts = Counter(cleanedList)

                #letter_counts = Counter(row)
                #print(type(letter_counts))
                most_com = OrderedDict(letter_counts.most_common(10))
                #most_com.pop('nan')
                

                #.most_common(10)
                df = pd.DataFrame.from_dict(most_com, orient='index')
                #df = to_dict_dropna(df_temp)
                if len(df):
                    df.plot(kind='bar')
                    plt.title(i + " Histogram")
                    plt.xlabel("Word")
                    plt.ylabel("Frequency")
                    plt.show()
                print('-' * 60)
        print "\t*******************************************"
        num = num + 1
    print "================================================="
    da_index = da_index + 1
```
